```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
rando_bi2= pd.read_csv("C:\\Users\\Ian\\OneDrive - Northeastern University\\ONE\\GIS_6345\\GIS6345_HW_WIP\\G6345_HW_WK8\\DATA\\bivariate_dataset.csv")

df_rando_bi2=pd.DataFrame(data=rando_bi2)

df_rando_bi2.head(5)
```

```python
x= df_rando_bi2['X'].values
y= df_rando_bi2['Y'].values
```

```python
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.3)
ax.scatter(x, y, s= 10, marker ='x', color='tab:green')
```
